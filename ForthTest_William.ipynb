```python
import statsmodels.api as sm

# Drop rows with NaN values from df_daily_returns before defining Y and X
df_daily_returns_cleaned_for_ols = df_daily_returns.dropna()

# Flatten the MultiIndex columns to a single level
df_daily_returns_cleaned_for_ols.columns = df_daily_returns_cleaned_for_ols.columns.get_level_values(0)

# Define dependent and independent variables from the cleaned DataFrame
Y = df_daily_returns_cleaned_for_ols['ANTM']
X = df_daily_returns_cleaned_for_ols[['BNB-USD', 'CL=F']]

# Add a constant (intercept) to the independent variables
X = sm.add_constant(X)

# Fit the OLS model with HC1 covariance type
model = sm.OLS(Y, X)
results = model.fit(cov_type='HC1')

# Display the regression results summary
print(results.summary())
```

```python
import yfinance as yf
import pandas as pd

# Define the tickers and date range
tickers = {'ANTM.JK': 'ANTM', 'BNB-USD': 'BNB-USD', 'CL=F': 'CL=F'}
start_date = '2022-01-01'
end_date = '2024-12-31'

# Dictionary to store the closing prices
closing_prices = {}

# Fetch data for each ticker
for yahoo_ticker, label in tickers.items():
    print(f"Fetching data for {label} ({yahoo_ticker})...")
    data = yf.download(yahoo_ticker, start=start_date, end=end_date)
    if not data.empty:
        closing_prices[label] = data['Close']
    else:
        print(f"No data found for {label} ({yahoo_ticker})")

# Combine all closing prices into a single DataFrame
if closing_prices:
    df_closing = pd.concat(closing_prices.values(), axis=1, keys=closing_prices.keys())
    df_closing.index.name = 'Date'
    print("\nHistorical Closing Prices:")
    display(df_closing.head())
    display(df_closing.tail())
else:
    print("No data was retrieved for any of the specified tickers.")
```
